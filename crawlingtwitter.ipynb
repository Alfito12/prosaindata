## 2. Import Library Twint

```python
import twint
import pandas
import nest_asyncio
nest_asyncio.apply()
```

## 3. Crawling data twitter dengan Twint

```python
c = twint.Config()
c.Search = "sandiagauno"
c.Since = "2023-01-05"
c.Until = "2023-02-22"
c.Limit = 200
c.Pandas = True
c.Lang = "id"

twint.run.Search(c)
```

* DataFrame hasil crawling twitter

```python
data = twint.storage.panda.Tweets_df
df = pandas.DataFrame(data, columns=['date', 'username', 'tweet'])
df
```
